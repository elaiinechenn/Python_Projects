```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
import matplotlib.pyplot as plt
import os
from IPython.display import Image as im
import csv
```

```python
r = requests.get('https://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=466920&stname=%25E8%2587%25BA%25E5%258C%2597&datepicker=2021-01-01')
#         https://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=466910&stname=%25E9%259E%258D%25E9%2583%25A8&datepicker=2021-01-01
```

```python
# 找台北所有場站名稱
soup = BeautifulSoup(r.text)
j = []
for i in range(20):
  station = soup.select('#selectStno')[0].text
  j.append(station)
print(j)
```

```python
s = soup.select('#selectStno')[0].text.split('\n')
s
```

```python
station_num = []
for station in s[1:-7]:
#  print(station.split('_')[0])
#  print(station.split('_')[1])
  station_num.append(station.split('_')[0])
print(station_num)
```

```python
station_name = []
for station in s[1:-7]:
    station_name.append(station.split('_')[1])
print(station_name)
```

```python
# 找payload裡的stname list
m=[]
for station in station_name:
 # print(station)
  w = requests.utils.quote(station.encode('utf8'))
# print(w)
  m.append(w)
#print(m)
str_ = " ".join(m)
print(str_)
```

```python
list_1 = []
str_1 = str_.replace('%','%25')
list_1.append(str_1)
#list_1.split( )
#print(list_1)
stname_1 = str_1.split(" ")
stname_1
```

```python
# 找日期
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2021, 1, 1)
end_date = date(2021, 3, 1)
datepicker_1 = []
for single_date in daterange(start_date, end_date):
    datepicker_1.append(single_date.strftime("%Y-%m-%d"))
print(datepicker_1)
```

```python
#更改網址內容
web_list=[]
for station,stname in zip(station_num,stname_1):
  a = (f"https://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station={station}&stname={stname}&datepicker=")
  web_list.append(a)
print(web_list)
```

```python
#加上日期
url_list=[]
for i in web_list:
  for datepicker in datepicker_1:
    adress = (f'{i}{datepicker}')
    url_list.append(adress)
url_list
```
